```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
from scipy import stats

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_mouse_df = pd.merge(study_results,mouse_metadata,how='outer', on=["Mouse ID"])

# Display the data table for preview
combined_mouse_df.head()
```

```python
# Checking the number of mice.
mice_count = combined_mouse_df["Mouse ID"].unique().size
print(f"Number of mice:{mice_count}")
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.

duplicate_mice = combined_mouse_df.loc[combined_mouse_df.duplicated(subset=['Mouse ID', 'Timepoint']), 'Mouse ID'].unique()
print(f"The duplicate mice by ID number: {duplicate_mice}")
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

duplicate_mice_data = combined_mouse_df[combined_mouse_df.duplicated(subset=["Mouse ID", "Timepoint"], keep=False)]
duplicate_mice_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

# Filter for rows with duplicate mouse IDs and timepoints
duplicate_rows = combined_mouse_df.drop_duplicates(subset=["Mouse ID", "Timepoint"], keep=False)
duplicate_rows

# Create a clean DataFrame by dropping duplicates (including the first occurrence)
clean_combined_mouse_df = combined_mouse_df.loc[combined_mouse_df['Mouse ID']!='g989']
clean_combined_mouse_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
new_mice_count = len(pd.unique(clean_combined_mouse_df["Mouse ID"]))
#len(mice_total_clean)
print(f"Number of mice:{new_mice_count} in the clean DataFrame")
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
#group DataFrame per regimen columns
grouped_df = clean_combined_mouse_df.groupby('Drug Regimen')

#calculate summary statistics
mean_tumor_volume = grouped_df['Tumor Volume (mm3)'].mean()
median_tumor_volume = grouped_df['Tumor Volume (mm3)'].median()
tumor_volume_variance = grouped_df['Tumor Volume (mm3)'].var()
tumor_volume_std = grouped_df['Tumor Volume (mm3)'].std()
tumor_volume_SEM = grouped_df['Tumor Volume (mm3)'].sem()

# place above statistics into a dataframe
summary_stats_df = pd.DataFrame({'Mean Tumor Volume': mean_tumor_volume, 
                                'Median Tumor Volume': median_tumor_volume,
                                'Tumor Volume Variance': tumor_volume_variance,
                                'Tumor Voume Std. Dev.': tumor_volume_std,
                                'Tumor Volume Std. Err.': tumor_volume_SEM})
summary_stats_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
#group DataFrame per regimen columns
grouped_df = clean_combined_mouse_df.groupby('Drug Regimen')
#calculate summary statistics
summary_stats = grouped_df['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
#Rename columns for clarity
summary_stats.columns = ['Mean Tumor Volume', 'Median Tumor Volume', 'Tumor Volume Variance', 'Tumor Voume Std. Dev.', 'Tumor Volume Std. Err.']
summary_stats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

#collect data for each drug from the cleaned dataframe
Capomulin_df = clean_combined_mouse_df.loc[clean_combined_mouse_df['Drug Regimen'] == 'Capomulin',:]
Ramicane_df = clean_combined_mouse_df.loc[clean_combined_mouse_df['Drug Regimen'] == 'Ramicane',:]
Infubinol_df = clean_combined_mouse_df.loc[clean_combined_mouse_df['Drug Regimen'] == 'Infubinol',:]
Ceftamin_df = clean_combined_mouse_df.loc[clean_combined_mouse_df['Drug Regimen'] == 'Ceftamin',:]

# Start by getting the last (greatest) timepoint for each mouse
Capomulin_last_timepoint = capomulin_df.groupby(['Drug Regimen', 'Mouse ID']).max()['Timepoint']
Capomulin_last_timepoint_df = pd.DataFrame(Capomulin_last_timepoint)
Capomulin_last_timepoint_df.head()

Ramicane_last_timepoint = Ramicane_df.groupby(['Drug Regimen', 'Mouse ID']).max()['Timepoint']
Ramicane_last_timepoint_df = pd.DataFrame(Ramicane_last_timepoint)
Ramicane_last_timepoint_df.head()

Infubinol_last_timepoint = Infubinol_df.groupby(['Drug Regimen', 'Mouse ID']).max()['Timepoint']
Infubinol_last_timepoint_df = pd.DataFrame(Infubinol_last_timepoint)
Infubinol_last_timepoint_df.head()

Ceftamin_last_timepoint = Ceftamin_df.groupby(['Drug Regimen', 'Mouse ID']).max()['Timepoint']
Ceftamin_last_timepoint_df = pd.DataFrame(Ceftamin_last_timepoint)
Ceftamin_last_timepoint_df.head()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
Capomulin_max_merge = pd.merge(Capomulin_last_timepoint, clean_combined_mouse_df, on=('Mouse ID', 'Timepoint'), how='left')
Capomulin_max_merge.head()

Ramicane_max_merge = pd.merge(Ramicane_last_timepoint, clean_combined_mouse_df, on=('Mouse ID', 'Timepoint'), how='left')
Ramicane_max_merge.head()

Infubinol_max_merge = pd.merge(Infubinol_last_timepoint, clean_combined_mouse_df, on=('Mouse ID', 'Timepoint'), how='left')
Infubinol_max_merge.head()

Ceftamin_max_merge = pd.merge(Ceftamin_last_timepoint, clean_combined_mouse_df, on=('Mouse ID', 'Timepoint'), how='left')
Ceftamin_max_merge.head()
```

```python
# A SIMPLER METHOD OF CALCULATING FINAL TUMOR VOLUME
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
greatest_timepoint = clean_combined_mouse_df.groupby(['Drug Regimen', 'Mouse ID'])['Timepoint'].max()
greatest_timepoint

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
group_max_merge = pd.merge(greatest_timepoint, clean_combined_mouse_df, on=('Mouse ID', 'Timepoint'), how='left')
group_max_merge.rename(columns={'Tumor Volume (mm3)': 'Last Tumor Volume (mm3)',
                              'Timepoint': 'Last Timepoint'}, inplace=True)
group_max_merge.head()

#clean data for specific treatment regimens:
#drop the unwanted treatment regimens so as to remain with Capomulin, Ramicane, Infubinol, and Ceftamin
clean_group_max_merge = group_max_merge.loc[group_max_merge['Drug Regimen'].isin(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])]
clean_group_max_merge.head()
```

```python
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
for treatment in treatments:
    print(treatment)
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

for treatment in treatments:
    
    # Create empty list to fill with tumor vol data (for plotting)
    tumor_vol = []

    # Calculate the IQR and quantitatively determine if there are any potential outliers.
    # I used loc to filter the DataFrame based on drug regimen, then accessed the 'Last Tumor Volume (mm3)'
    quartiles = clean_group_max_merge.loc[clean_group_max_merge['Drug Regimen']==treatment]['Last Tumor Volume (mm3)'].quantile([.25,.5,.75])
    Q1 = quartiles[.25]
    Q3 = quartiles[.75]
    IQR = Q3 - Q1
    
    print(f"-----------------------------------------------------------------------")
    print(f"The first quartile (Q1) of {treatment} is: {Q1}")
    print(f"The third quartile (Q3) of {treatment} is: {Q3}")
    print(f"The Interquartile Range (IQR) of {treatment} is: {IQR}")   

# Locate the rows which contain mice on each drug and get the tumor volumes

    tumor_volume_df = clean_group_max_merge.loc[:, ['Drug Regimen', 'Last Tumor Volume (mm3)']]
    tumor_volume_df
    
    # add subset 
tumor_vol.append(tumor_volume_df)
    
    # Determine outliers using upper and lower bounds
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)
print(f"Values below {lower_bound} could be {treatment} outliers.")
print(f"Values above {upper_bound} could be {treatment} outliers.")
print(f"-------------------------------------------------------------------------")
```

```python
# Using loc function to select specific colums 'Drug Regiment' and 'Last Tumor Volume (mm3) from the clean merged DataFrame 
tumor_volume_df = clean_group_max_merge.loc[:, ['Drug Regimen', 'Last Tumor Volume (mm3)']]
print(tumor_volume_df)
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
tumor_vol = []

# Appending tumor volume data to the list
for treatment in treatments:
    tumor_vol.append(tumor_volume_df.loc[tumor_volume_df['Drug Regimen'] ==treatment, 'Last Tumor Volume (mm3)']) 

# Create a boxplot
plt.figure(figsize=(10,8))
plt.boxplot(tumor_vol, labels=treatments, patch_artist=True)
plt.ylabel('Last Tumor Volume (mm3)')
plt.xlabel('Drug Regimen')
plt.title('Distribution of Last Tumor Volume (mm3) per Treatment Group', fontsize=18)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
# Filter dataset to include only the mice treated with Capomulin
Capomulin_df = clean_combined_mouse_df.loc[clean_combined_mouse_df['Drug Regimen'] == 'Capomulin',:]

# Choose a specific mouse from the filtered data above
single_mouse = Capomulin_df['Mouse ID'].unique()[0]
print(single_mouse)

# Filter out the tumor volume data for the above selected mouse
single_mouse_data = Capomulin_df[Capomulin_df['Mouse ID'] == single_mouse]

# Plot line graph for the specific mouse
plt.plot(single_mouse_data['Timepoint'], single_mouse_data['Tumor Volume (mm3)'], marker='*', color='blue')
plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Mouse(b128) Tumor Volume in Capomulin Treatment', fontweight='bold')
plt.show()
```

```python
# confirm that Capomulin_df has the data we need to generate scatter plot
Capomulin_df = clean_combined_mouse_df.loc[clean_combined_mouse_df['Drug Regimen'] == 'Capomulin',:]
Capomulin_df.head()
```

```python
# confirm that i have the right data (avg. weight & tumor volume) needed for the scatterplot below
Capomulin_gp = Capomulin_df.groupby('Mouse ID')[['Weight (g)','Tumor Volume (mm3)']].mean()
Capomulin_gp.rename(columns={'Tumor Volume (mm3)': 'Avg Tumor Volume (mm3)',
                              'Weight (g)': 'Avg Weight (g)'}, inplace=True)
Capomulin_gp
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
# I shall use the same mouse ID as above (in the line plot)
single_mouse = Capomulin_df['Mouse ID'].unique()[0]

# calculate the average weight and observed tumor volume for the entire Capomulin treatment
Capomulin_df = clean_combined_mouse_df.loc[clean_combined_mouse_df['Drug Regimen'] == 'Capomulin',:]
Capomulin_gp = Capomulin_df.groupby('Mouse ID')[['Weight (g)','Tumor Volume (mm3)']].mean()


# create scatter plot
plt.scatter(Capomulin_gp['Weight (g)'], Capomulin_gp['Tumor Volume (mm3)'])
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Relation Between Mouse Weight & Tumor Volume in Capomulin Regimen', fontweight='bold')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient for mouse weight and average observed tumor volume for the entire Capomulin regimen
x_values = Capomulin_gp['Weight (g)']
y_values = Capomulin_gp['Tumor Volume (mm3)']
# Calculate the correlation coefficient
correlation_coeffficient = np.corrcoef(Capomulin_gp['Weight (g)'], Capomulin_gp['Tumor Volume (mm3)'])[0][1] # # Access row 0, column 1 from the first element

print(f"The Correlation coefficient (Pearson's R) between mouse weight and average tumor volume is {correlation_coeffficient}")
```

```python
# Calculate linear regression values
regress_values = stats.linregress(Capomulin_gp['Weight (g)'], Capomulin_gp['Tumor Volume (mm3)'])
print(regress_values)
```

```python
# Calculate the a linear regression model for mouse weight and average observed tumor volume for the entire Capomulin regimen
# Calculate linear regression values
regress_values = st.linregress(Capomulin_gp['Weight (g)'], Capomulin_gp['Tumor Volume (mm3)'])

# equation of line calculation (y = mx +b)
line_eq = "y = " + str(round(regress_values.slope, 2)) + "x + " + str(round(regress_values.intercept, 2))

# create scatter plot with linear regression line, equation of line , & r-squared value
plt.scatter(Capomulin_gp['Weight (g)'], Capomulin_gp['Tumor Volume (mm3)'])
plt.plot(Capomulin_gp['Weight (g)'], regress_values.slope * Capomulin_gp['Weight (g)'] + regress_values.intercept, color='r')
plt.annotate(line_eq, (20,36),fontsize=10, color='r')
plt.text (20, 35, f'R-squared: {regress_values.rvalue**2}', fontsize=10, color='r') 
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Relation Between Mouse Weight & Tumor Volume in Capomulin Regimen', fontweight='bold')
plt.show()
```
